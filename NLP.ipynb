```python
import pandas as pd
```

```python
df=pd.read_csv('/content/vaccination_tweets.csv')
```

```python
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.tokenize import word_tokenize , wordpunct_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
df.columns
```

```python
# removing unwanted columns
df1=df.drop(['id', 'user_name', 'user_location', 'user_description', 'user_created',
       'user_followers', 'user_friends', 'user_favourites', 'user_verified',
       'date', 'hashtags', 'source', 'retweets', 'favorites',
       'is_retweet'],axis = 1)
df1
```

```python
def data_processing(text):

    #Converting into lowecase
    text = text.lower()

    #Removing with Urls
    text = re.sub(r"https\S+|www\S+https\S+", '',text, flags=re.MULTILINE)

    #handling Punctuations
    text = re.sub(r'\@w+|\#','',text)
    text = re.sub(r'[^\w\s]','',text)

    #removing the stopwords
    text_tokens = word_tokenize(text)
    filtered_text = [w for w in text_tokens if not w in stop_words]

    return " ".join(filtered_text)

df1.text = df1['text'].apply(data_processing)
```

```python
print("length of dataset",len(df1))
```

```python
df2=df1.drop_duplicates('text')
print("lenght of dataset after removing duplicates",len(df2))
```

```python
# Stemming
stemmer = PorterStemmer()
def stemming(data):
    text = [stemmer.stem(word) for word in data]
    return data
df2['text'] = df2['text'].apply(lambda x: stemming(x))
```
